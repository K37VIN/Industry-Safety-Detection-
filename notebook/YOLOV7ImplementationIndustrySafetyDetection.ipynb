```python
# Prepare img path in text file
train_img_path= "/content/drive/MyDrive/ISD/images/train"
test_img_path= "/content/drive/MyDrive/ISD/images/val"
```

```python
import os
```

```python
#Training images
with open('train.txt',"a+") as f:
  img_list = os.listdir(train_img_path)
  for img in img_list:
    f.write(os.path.join(train_img_path,img+'\n'))
  print("Done")
```

```python
# Validation images
with open('val.txt',"a+") as f:
  img_list = os.listdir(test_img_path)
  for img in img_list:
    f.write(os.path.join(test_img_path,img+'\n'))
  print("Done")
```
